```python
# Crear dos personajes de una historia que se representen a traves de clases
# Cada personaje debe de tener poderes o acciones diferentes
# Deben presentarse en una frase


# Clase madre  del Personaje
class Personaje:

    tipo = 'Misterioso'
    def __init__(self, nombre):
        self.nombre = nombre

    def presentarse(self):
        pass  


# Crear clase hija
class Heroe(Personaje):
    def __init__(self, nombre, poder):
        super().__init__(nombre)
        self.poder = poder
    # Imprimir la presentacion 
    def presentarse(self):
        print(f"Mi nombre es {self.nombre} y poseo la habilidad de {self.poder}")


# Crear clase hija del villano
class Villano(Personaje):
    def __init__(self, nombre, habilidad):
        super().__init__(nombre)
        self.habilidad = habilidad

    def presentarse(self):
        print(f"Mi nombre es {self.nombre} y pongo {self.habilidad}")


# Crear los objetos
heroe = Heroe("thiaguito", "cagadera extrema")
villano = Villano("Penguin","Challenges dificiles, super dificiles >:p")

# ahora las presentaciones
heroe.presentarse()
villano.presentarse()

    
```
