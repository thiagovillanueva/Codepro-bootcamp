```python
# Elegir una contrasenha
contrasenha = 9876
#colocar while y designarle True para empezar
while True:

    # Hacer la pregunta al usuario e pedir contrasenha
    contrasenha_a_ingresar = int(input('Ingrese la contrasenha por favor'))
    if contrasenha_a_ingresar == contrasenha:
       # Si la contrasenha es correcta se mostrara esto:
        print('contrasenha correcta')
        break
    else:
       # Si la contrasenha es incorrecta se muestra esto:
        print('La contrasenha es incorrecta intentelo de nuevo')

# Mensaje final una vez ingresado correctamente
print('bienvenido sr:D')

```
